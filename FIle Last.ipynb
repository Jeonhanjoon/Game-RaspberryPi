```python
import time
import random
from colorsys import hsv_to_rgb
import board
from digitalio import DigitalInOut, Direction
from PIL import Image, ImageDraw, ImageFont
from adafruit_rgb_display import st7789
import numpy as np  
import os
```

```python

class Joystick:
    def __init__(self):
        self.cs_pin = DigitalInOut(board.CE0)
        self.dc_pin = DigitalInOut(board.D25)
        self.reset_pin = DigitalInOut(board.D24)
        self.BAUDRATE = 24000000

        self.spi = board.SPI()
        self.disp = st7789.ST7789(
                    self.spi,
                    height=240,
                    y_offset=80,
                    rotation=180,
                    cs=self.cs_pin,
                    dc=self.dc_pin,
                    rst=self.reset_pin,
                    baudrate=self.BAUDRATE,
                    )

        # Input pins:
        self.button_A = DigitalInOut(board.D5)
        self.button_A.direction = Direction.INPUT

        self.button_B = DigitalInOut(board.D6)
        self.button_B.direction = Direction.INPUT

        self.button_L = DigitalInOut(board.D27)
        self.button_L.direction = Direction.INPUT

        self.button_R = DigitalInOut(board.D23)
        self.button_R.direction = Direction.INPUT

        self.button_U = DigitalInOut(board.D17)
        self.button_U.direction = Direction.INPUT

        self.button_D = DigitalInOut(board.D22)
        self.button_D.direction = Direction.INPUT

        self.button_C = DigitalInOut(board.D4)
        self.button_C.direction = Direction.INPUT

        # Turn on the Backlight
        self.backlight = DigitalInOut(board.D26)
        self.backlight.switch_to_output()
        self.backlight.value = True

        # Create blank image for drawing.
        # Make sure to create image with mode 'RGB' for color.
        self.width = self.disp.width
        self.height = self.disp.height
```

```python


import numpy as np

# Constants
SCREEN_WIDTH = 240
SCREEN_HEIGHT = 240

class Dinosaur:
    def __init__(self, width, height):
        self.state = None
        self.position = np.array([width/2 - 20, height/2 - 20, width/2 + 20, height/2 + 20])
    
        self.center = np.array([(self.position[0] + self.position[2]) / 2, (self.position[1] + self.position[3]) / 2])
        self.outline = "#FFFFFF"
        self.duck_img = DUCKING
        self.run_img = RUNNING
        self.jump_img = JUMPING

        self.dino_duck = False
        self.dino_run = True
        self.dino_jump = False

        self.image = self.run_img[0]
        self.dino_rect = self.image.get_rect()
        self.dino_rect.x = self.position[0]
        self.dino_rect.y = self.position[1]

    def update(self, command):
        if command['down_pressed']:
            self.dino_duck = True
            self.dino_run = False
            self.dino_jump = False
        elif command['up_pressed']:
            self.dino_duck = False
            self.dino_run = False
            self.dino_jump = True
        elif command['move']:
            self.dino_duck = False
            self.dino_run = True
            self.dino_jump = False

        if self.dino_duck:
            self.duck()
        if self.dino_run:
            self.run()
        if self.dino_jump:
            self.jump()

        self.move(command)

    def move(self, command):
        if command['move'] == False:
            self.state = None
            self.outline = "#FFFFFF"
        
        else:
            self.state = 'move'
            self.outline = "#FF0000"

            if command['up_pressed']:
                self.position[1] -= 5
                self.position[3] -= 5

            if command['down_pressed']:
                self.position[1] += 5
                self.position[3] += 5

            if command['left_pressed']:
                self.position[0] -= 5
                self.position[2] -= 5
                
            if command['right_pressed']:
                self.position[0] += 5
                self.position[2] += 5
                
        #center update
        self.center = np.array([(self.position[0] + self.position[2]) / 2, (self.position[1] + self.position[3]) / 2])

    def duck(self):
        self.image = self.duck_img[0]
        self.dino_rect = self.image.get_rect()
        self.dino_rect.x = self.position[0]
        self.dino_rect.y = self.position[1]

    def run(self):
        self.image = self.run_img[0]
        self.dino_rect = self.image.get_rect()
        self.dino_rect.x = self.position[0]
        self.dino_rect.y = self.position[1]

    def jump(self):
        self.image = self.jump_img
        if self.dino_jump:
            self.position[1] += 5
            self.position[3] += 5

    def draw(self, SCREEN):
        SCREEN.blit(self.image, (self.dino_rect.x, self.dino_rect.y))
png_image = Image.open('/home/hjun1/esw/TA-ESW/1.Basic/STARTBACKGROUND-2.png')
```

```python

class Character:
    def __init__(self, width, height):
        self.image = Image.open("/home/hjun1/esw/TA-ESW/1.Basic/STOP.png")
        self.state = None
        self.position = np.array([width/2 - 20, height/2 - 20, width/2 + 20, height/2 + 20])
        self.outline = "#FFFFFF"
        self.image = Image.open("/home/hjun1/esw/TA-ESW/1.Basic/STOP.png")
        datas = self.image.getdata()
        new_data = []
        for item in datas:
            if item[0] > 200 and item[1] > 200 and item[2] > 200:
                new_data.append((255, 255, 255, 0))
            else:
                new_data.append(item)
        self.image.putdata(new_data)
        self.is_jumping = False # 점프 중인지 나타내는 상태 변수 추가
        self.jump_height = 10 # 점프 높이 

    def jump(self):
        if self.is_jumping:
            return
        self.is_jumping = True
        self.position[1] -= self.jump_height
        self.position[3] -= self.jump_height

    def land(self):
        if not self.is_jumping:
            return
        self.is_jumping = False
        self.position[1] += self.jump_height
        self.position[3] += self.jump_height

    def move(self, command = None, background=None):
        if command and 'move' in command and not command['move']:
            self.state = None 
       
        else:
            self.state = 'move'
            self.outline = "#FF0000"

            if command['up_pressed']is not None and command['up_pressed']:
                self.position[1] += 5
                self.position[3] += 5

            if command['down_pressed']:
                self.position[1] += 0
                self.position[3] += 0

            if command['left_pressed']:
                self.position[0] -= 0
                self.position[2] -= 0
                
            if command['right_pressed']:
                self.position[0] += 5
                self.position[2] += 5

            if command['jump_pressed'] and not self.is_jumping:
                self.jump()

            if not command['jump_pressed'] and self.is_jumping:
                self.land()
my_image = Image.new("RGBA", (joystick.width, joystick.height), (135, 206, 235, 0))
```

```python
import tkinter as tk
from threading import Thread
import random

SCREEN_WIDTH = 240
SCREEN_HEIGHT = 240

class Cloud:
    def __init__(self, canvas):
        self.canvas = canvas
        self.x = SCREEN_WIDTH // 2
        self.y = SCREEN_HEIGHT // 2
        self.image = canvas.create_oval(self.x - 10, self.y - 10, self.x + 10, self.y + 10, fill="blue")

    def update(self, direction):
        speed = 5
        if direction == "up" and self.y - speed >= 0:
            self.y -= speed
        elif direction == "down" and self.y + speed <= SCREEN_HEIGHT:
            self.y += speed
        elif direction == "left" and self.x - speed >= 0:
            self.x -= speed
        elif direction == "right" and self.x + speed <= SCREEN_WIDTH:
            self.x += speed

        self.canvas.coords(self.image, self.x - 10, self.y - 10, self.x + 10, self.y + 10)

class JoystickApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Joystick App")
        self.root.geometry(f"{SCREEN_WIDTH}x{SCREEN_HEIGHT}")

        self.canvas = tk.Canvas(self.root, width=SCREEN_WIDTH, height=SCREEN_HEIGHT, bg="white")
        self.canvas.pack()

        self.cloud = Cloud(self.canvas)

        self.joystick_thread = Thread(target=self.update_joystick)
        self.joystick_thread.start()

    def update_joystick(self):
        while True:
            # Replace this part with your actual joystick input handling
            direction = input("Enter direction (up/down/left/right): ")
            self.cloud.update(direction)
            self.root.update_idletasks()
            self.root.update()
```
